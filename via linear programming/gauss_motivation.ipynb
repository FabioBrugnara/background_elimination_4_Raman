# The translated gaussian functions are limited in frequency

```python
gauss = lambda z,a,K: np.exp(-4*np.pi**2*K*(z-a)**2)
Fgauss = lambda f,a,K: np.exp(-1j*2*np.pi*f*a)*np.exp(-f**2/(4*K))/np.sqrt(4*np.pi*K)
```

```python
K=10
a=20
f=np.arange(0,80)
plt.figure(figsize=(10,5))
plt.plot(f, np.abs(Fgauss(f,a,K)))
#plt.plot(f, (gauss(f,a,K)))
```
